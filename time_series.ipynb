```python
filtered_columns = [col for col in df_world.columns if 'consumption' in col.lower()]
filtered_columns.append('year')
strings_to_remove = ['primary_energy_consumption']
filtered_columns = list(filter(lambda item: item not in strings_to_remove,filtered_columns))
df_filtered = df_world[filtered_columns]
```

```python
import pandas as pd
import plotly.graph_objects as go


renewable_cols = ['hydro_consumption', 'solar_consumption', 'wind_consumption', 'biofuel_consumption', 'other_renewable_consumption', 'renewables_consumption']
non_renewable_cols = ['coal_consumption', 'gas_consumption', 'oil_consumption', 'fossil_fuel_consumption']
nuclear_cols = ['nuclear_consumption']

# Sort by year and sum the consumption per category per year
df = df_world.sort_values(by='year')
df['Renewable'] = df[renewable_cols].sum(axis=1)
df['Non_Renewable'] = df[non_renewable_cols].sum(axis=1)
df['Nuclear'] = df[nuclear_cols].sum(axis=1)

# Create a Plotly graph
fig = go.Figure()

# Adding traces
fig.add_trace(go.Scatter(x=df['year'], y=df['Renewable'], mode='lines+markers', name='Renewable Energy Consumption', line=dict(color='green')))
fig.add_trace(go.Scatter(x=df['year'], y=df['Non_Renewable'], mode='lines+markers', name='Non-Renewable Energy Consumption', line=dict(color='darkred')))
fig.add_trace(go.Scatter(x=df['year'], y=df['Nuclear'], mode='lines+markers', name='Nuclear Energy Consumption', line=dict(color='blue')))

# Layout settings
fig.update_layout(
    title='Energy Consumption Over Years',
    xaxis_title='Year',
    yaxis_title='Energy Consumption - Measured in terawatt-hours',
    legend_title='Energy Types',
    template='plotly_white'
)

# Save the plot as an HTML file
fig.write_html('plot_aggregated_world.html')
```

```python
import pandas as pd
import plotly.graph_objects as go

# Assuming df_world is pre-loaded with the correct data
# Define columns for different energy types
renewable_cols = ['hydro_consumption', 'solar_consumption', 'wind_consumption', 'biofuel_consumption', 'other_renewable_consumption', 'renewables_consumption']
non_renewable_cols = ['coal_consumption', 'gas_consumption', 'oil_consumption', 'fossil_fuel_consumption']
nuclear_cols = ['nuclear_consumption']

# Sort by year and filter the DataFrame to include only data after 1950
df_world.sort_values(by='year', inplace=True)
df_filtered = df_world[df_world['year'] > 1950].copy()  # Use .copy() to avoid SettingWithCopyWarning

# Sum the consumption for each energy category using .loc to safely assign values
df_filtered.loc[:, 'Renewable'] = df_filtered[renewable_cols].sum(axis=1)
df_filtered.loc[:, 'Non_Renewable'] = df_filtered[non_renewable_cols].sum(axis=1)
df_filtered.loc[:, 'Nuclear'] = df_filtered[nuclear_cols].sum(axis=1)

# Create a Plotly graph
fig = go.Figure()

# Adding traces for each energy type
fig.add_trace(go.Scatter(
    x=df_filtered['year'], 
    y=df_filtered['Renewable'], 
    mode='lines+markers', 
    name='Renewable Energy Consumption', 
    line=dict(color='green')
))
fig.add_trace(go.Scatter(
    x=df_filtered['year'], 
    y=df_filtered['Non_Renewable'], 
    mode='lines+markers', 
    name='Non-Renewable Energy Consumption', 
    line=dict(color='red')
))
fig.add_trace(go.Scatter(
    x=df_filtered['year'], 
    y=df_filtered['Nuclear'], 
    mode='lines+markers', 
    name='Nuclear Energy Consumption', 
    line=dict(color='blue')
))

# Layout settings
fig.update_layout(
    title='Energy Consumption Over Years ',
    xaxis_title='Year',
    yaxis_title='Energy Consumption - Measured in terawatt-hours',
    legend_title='Energy Types',
    template='plotly_white'
)

# Save the plot as an HTML file
fig.write_html('./outputs/plot_aggregated_world.html')

```
